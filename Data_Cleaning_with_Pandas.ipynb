```python
import pandas as pd
```

```python
df = pd.read_excel('Customer Call List.xlsx')
```

```python
df
```

```python


df = df.drop_duplicates()
df

```

```python

df = df.drop(columns = 'Not_Useful_Column')
df
```

```python

df['Last_Name'].str.strip('/..._')
```

```python

df['Last_Name'] = df['Last_Name'].str.strip('/..._')
df
```

```python

df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '')
```

```python

df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex = True)
df
```

```python
#To convert Phone Numbers data type to int.
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:])
```

```python

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:])
df
```

```python
df['Phone_Number'] = df['Phone_Number'].replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].replace('Na--', '')
df
```

```python

df['Address'].str.split(',', 2, expand = True)
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', 2, expand = True)
df
```

```python

df['Paying Customer'].str.replace('Yes', 'Y')
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')



df
```

```python

df = df.replace('N/a', '')
df = df.fillna('')
df
```

```python
df = df.rename(columns={"Stree_Address": "Street_Address"})
df
```

```python

for x in df.index:
  if df.loc[x, "Do_Not_Contact"] ==  'Y':
    df.drop(x, inplace = True)

df
```

```python

for x in df.index:
  if df.loc[x, 'Phone_Number'] == '':
    df.drop(x, inplace = True)

df
```
