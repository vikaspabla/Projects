# Importing Necessary Libraries

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Reading csv file and assigning it to dataframe

```python
df = pd.read_csv("world_population.csv")
df
```

```python
#Chnaging float numbers to 2 decimal points
pd.set_option('display.float_format', '{:.2f}'.format)
```

# An overhead view of the dataframe

```python
df.info()
```

# Returning descriptionof the dataframe to look at various values like mean, min and max.

```python
df.describe()
```

# Checking the dataframe for null values

```python
df.isnull().sum()
```

# Looking at unique values in the dataframe

```python
df.nunique()
```

# Sorting the dataframe with the largest poulation of the latest year

```python
df.sort_values(by = '2022 Population', ascending = False).head(10)
```

# Finding the correlation between the columns of dataframe

```python
df.corr(numeric_only = True)
```

# Visualizing Correlation in a HeatMap using Seaborn

```python
sns.heatmap(df.corr(numeric_only = True), annot = True)
plt.rcParams['figure.figsize'] = (10,7)
plt.show()
```

# Grouping the data by continents

```python
df.groupby('Continent').mean(numeric_only=True)
```

# Plotting a line graph of data grouped by continents with increasing years of poulation

```python
continents = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only=True).sort_values(by = '2022 Population', ascending = False)
continents.plot()
```

# Transposing data

```python
continents.transpose()
```

# Viasualizing transposed data of population by continents with a line graph

```python
transposed = continents.transpose()
plt.rcParams['figure.figsize'] = (15,7)
transposed.plot()
```

# Checking for Outliers

```python
df.boxplot(figsize = (25,10))
```

# Choosing only 2 columns to show in a bar graph

```python
df.iloc[:, [2,16]]
```

```python
countries = df.iloc[:, [2,16]].sort_values(by = 'World Population Percentage', ascending = False).head(10)
countries.groupby('Country')['World Population Percentage']
countries
```

```python
sns.barplot(data = countries, x = 'Country', y = 'World Population Percentage')
```
